```python
import pandas as pd
import numpy as np
import datacompy

data_season1 =pd.read_csv('peoplec.csv',sep=';')
data_season2 = pd.read_xml('peoplex.xml')
df1 = pd.DataFrame(data_season1)
df2 = pd.DataFrame(data_season2)
new_row1 = {'id':'5', 'name':'David', 'lastname':'Garcia', 'age':32,'weight':45, 'height':1.11, 'sex':'male', 'birthdate':'1990-11-22', 'birthplace':'Caracas'}
new_row2 = {'id':'5', 'name':'Esperanza', 'lastname':'Gaucho', 'age':23,'weight':45, 'height':1.80, 'sex':'female'}
df1=df1.append(new_row1, ignore_index=True)
df2=df2.append(new_row2, ignore_index=True)
```

```python
dif=datacompy.Compare(df1,df2,join_columns='id')
print(dif.df2_unq_rows)
print(dif.df1_unq_rows)
```
