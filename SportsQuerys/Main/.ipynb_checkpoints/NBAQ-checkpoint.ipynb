Queremos saber de cada universidad quién ha sido el jugador con más minutos jugados en una temporada

```python

dictUni={}
unis = []
jugs = []
for i in df_nba['Universidad']:
    if i not in unis:
        unis.append(i)
j=0
while (j<len(unis)):
  univ=df_nba.loc[:,'Universidad']==unis[j]
  universidad=df_nba.loc[univ]
  maxUni=universidad['Minutos_Jugados'].max()
  jugMax= universidad.loc[:,'Minutos_Jugados']==maxUni
  campeon=universidad.loc[jugMax]
  nombre=campeon['Jugador']
  for e in nombre:
    if e not in jugs:
        jugs.append(e)
  dictUni[j] = {unis[j]: jugs[e] for e in range(len(jugs))}
  j=j+1


print(dictUni)
```

```python
temp2014 = df_nba.loc[df_nba["Temporada"]==2014]
temp2014.head()
```

```python
#Sabiendo que Philadelphia fue el mejor equipo en 2014, ver en qué deben mejorar el resto de equipos

phi_tm = df_nba.loc[:, 'Equipo'] == 'PHI'
phi_df=df_nba.loc[phi_tm]
phi2014_tm = phi_df.loc[:, 'Temporada'] == 2014
phit2014_df = phi_df.loc[phi2014_tm]
phil=phit2014_df[['2P','3P','AST','ORB','DRB','BLK','STL']]
total=temp2014[['2P','3P','AST','ORB','DRB','BLK','STL']]
total.mean().plot(kind='bar',color="#69252f",figsize=(30,15))#en rojo la media de los equipos de la t2014
phil.mean().plot(kind='bar',color="#00ff73") #en verde la media de Philadelphia 76ers
plt.title("¿Qué hay que mejorar?")
plt.show()
```
